**JOINT PLOT**

```python
speeding_alcohol = df.groupby(['speeding', 'alcohol']).size().unstack()
speeding_alcohol.plot(kind='bar', stacked=True)
plt.xlabel("Speeding")
plt.ylabel("Count")
plt.title("Stacked Bar Chart of Speeding and Alcohol-Related Accidents")
plt.legend(title="Alcohol")
plt.xticks([0, 1], ['No Speeding', 'Speeding'])
plt.show()
# Create a stacked bar chart for 'speeding' and 'alcohol'
# Inference: This stacked bar chart shows the combined effect of speeding and alcohol-related accidents.
```

```python
sns.jointplot(x="total",y="speeding",data=df)
#This generates a joint plot using Seaborn, displaying a combination of scatter plot and histograms
#to explore the relationship between the "total" and "speeding" variables in the DataFrame "df." It helps visualize their distribution and correlation.
```

```python
sns.jointplot(x="alcohol",y="ins_losses",data=df,color="green")
#This code generates a joint plot using Seaborn, displaying a combination of scatter plot and histograms
#to explore the relationship between the "alcohol" and "ins_losses" variables in the DataFrame "df." It helps visualize their distribution and correlation.
```

**BOX PLOT**

```python
plt.subplots(figsize=(20,8))
sns.boxplot(x="total", y="no_previous", data=df)
plt.xlabel("Total Accidents")
plt.ylabel("Percentage of Drivers with No Previous Accidents")
plt.title("Boxplot of Total Accidents vs. No Previous Accidents")
plt.show()
#INFERENCE:
#The boxplot illustrates a negative relationship between "Total Accidents" and the "Percentage of Drivers with No Previous Accidents."
#As the total number of accidents increases, there is a tendency for the percentage of drivers with no previous accidents to decrease.
#This indicates an inverse proportionality between these variables, where higher accident rates are associated with a lower percentage of inexperienced drivers.
```

```python
plt.subplots(figsize=(20,5))
sns.boxplot(x="alcohol", y="no_previous", data=df)
#INFERENCE: INVERSLY PROPORTIONAL The boxplot between "alcohol" and "no_previous" in the car_crashes dataset shows a notable pattern.
#As alcohol involvement increases, the number of drivers with no previous accidents tends to decrease
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="not_distracted", y="total", data=df)
plt.xlabel("Distracted Driving")
plt.ylabel("Total Accidents")
plt.title("Box Plot of Total Accidents by Distracted Driving Status")
plt.show()
# Inference:The boxplot between "not_distracted" and "total" accidents in the car_crashes dataset reveals
# that regions with higher levels of not being distracted tend to have a wider range of total accidents.
```

**Heat Map**

```python
sns.heatmap(corr,annot=True)
#This creates a heatmap using Seaborn to visualize the correlation matrix ("corr") with annotated values,
#providing a graphical representation of the strength and direction of relationships between numerical columns in the DataFrame.
```

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
#Inference: In the heatmap above, blue denotes extreme numbers that are positively associated, while green shows values that are negatively correlated.
#We can get car wrecks more precisely since the faster the speed,
#the more likely it is to have an accident.It also indicates that alcohol consumption and carcrashes are directly proportional.
#It also reveals where values are when drivers are not distracted but have a car accident, where insurance premiums are not involved,
#and when losses are not implicated in the same way.Extreme values are shown in dark blue, whereas minimal values are shown in light green.
```
